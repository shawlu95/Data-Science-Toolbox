```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("datasets/titanic_data.csv")
print(df.shape)
df.head()
```

```python
for col in df.columns:
    if any(df[col].isnull()):
        num_missing = sum(df[col].isnull())
        print("feature %s, missing %i entries (%.2f%%)"
              %(col, num_missing, 100 * num_missing / len(df)))
    else:
        print("feature %s has no missing value"%col)
```

#### Drop rows that contain any null

```python
df.dropna(axis=0, how='any').shape
```

#### Drop rows that have particular columns set to null

```python
df.dropna(subset=["Age"]).shape
```

#### Drop entire column

```python
df.drop("Age", axis=1).shape
```
