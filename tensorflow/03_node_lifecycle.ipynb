```python
import tensorflow as tf
```

```python
w = tf.constant(3)

x = w + 2
y = x + 5
z = x * 3
```

```python
with tf.Session() as sess:
    # note that x is evaluated twice
    print(y.eval()) # 10
    print(z.eval()) # 15
```

```python
with tf.Session() as sess:
    # evaluate x once, in a single graph run
    y_val, z_val = sess.run([y, z])
print(y_val, z_val)
```
