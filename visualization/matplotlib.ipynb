```python
import numpy as np
from matplotlib import pyplot as plt
```

### Line plot

```python
plt.figure(figsize=(8, 4))

# create sample data
x = np.linspace(- np.pi, 2 * np.pi, 100)
sinx = np.sin(x)
cosx = np.cos(x)

# plot two lines
plt.plot(x, sinx)
plt.plot(x, cosx)

# plot verticalï¼Œ horizontal lines
plt.axhline(0, linestyle = "--", color = 'grey')
plt.axvline(0, linestyle = "--", color = 'grey')

plt.plot([0], [0], 'o', markersize=10, color='b')
plt.plot([np.pi], [np.cos(np.pi)], 'v', markersize=10, color='g')
plt.plot([np.pi/2], [np.sin(np.pi/2)], '^', markersize=10, color='r')

# annotation
plt.legend(["sine(x)", "cos(x)"])
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sample Plot")

plt.show()
```

### Scatter 

```python
plt.figure(figsize=(8, 4))

# create sample data
x1 = np.random.normal(10, 0.2, 1000)
y1 = np.random.normal(5, 1, 1000)

x2 = np.random.normal(8, 0.8, 1000)
y2 = np.random.normal(6, 0.3, 1000)

# set axis limit
plt.xlim([5, 12])
plt.ylim([0, 10])

# alternative: set axis as [x_low, x_high, y_low, y_high]
# plt.axis([5, 12, 0, 10])

plt.grid(True)

plt.scatter(x1, y1, alpha=0.3, marker="v")
plt.scatter(x2, y2, alpha=0.3, marker="o")

plt.show()
```

### Histogram

```python
plt.figure(figsize=(8, 4))

# create sample data
sigma = 1
mu = 5
x = np.random.normal(mu, sigma, 1000)

# divide into bins
count, bins, ignored = plt.hist(x, 30, density=True, facecolor = 'g', alpha = 0.4)

# plot histogram and probability density
plt.plot(bins, 
         1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)), 
         linewidth=2, color='r')

plt.grid(True)

# use math expression in title
plt.title('$\sigma_s=%i$, $\mu_s=%i$'%(sigma, mu))

plt.show()
```

```python
plt.figure(figsize=(8, 4))

# create sample data
x1 = np.random.normal(1, 1, 1000)
x2 = np.random.normal(5, 3, 1000)

# divide into bins
count, bins, ignored = plt.hist([x1, x2], 30, density = True, alpha = 0.4, label = ["x1", "x2"])

plt.grid(True)
plt.legend(loc='upper left')

plt.show()
```

### Pie chart

```python
plt.figure(figsize=(8, 8))

# values must be numbers, labels must be strings
values = np.linspace(0, 80, 5)
labels = [str(val) for val in values]

_ = plt.pie(x = values, 
            labels = labels,
            autopct='%1.1f%%',     # add percentage
            labeldistance = 1.1)   # percentage tag distance from center

# draw a circle at center
centre_circle = plt.Circle((0,0),0.70,fc='white')
plt.gca().add_artist(centre_circle)

plt.show()
```

### Subplot

```python
plt.figure(figsize=(8, 8))

# subplot setup
nplots = 4
ncols = 2
nrows = 2
colors = ['b', 'r', 'g', 'm']

# create fake data
mu = [0, 1, 4, 8]
sigma = [0.1, 3, 0.02, 10]
data = [np.random.normal(mu[i], sigma[i], 1000) for i in range(nplots)]

for i in range(nplots):
    x = data[i]

    # parameter: number of rows, number of columns, plot index
    ax = plt.subplot(nrows, ncols, i + 1)
    
    # can be any type of plot (histogram, pie, line etc.)
    ax.hist(x, 50, density=True, color=colors[i], alpha = 0.4)
    
    ax.set_ylabel("Density")
    ax.set_title("Plot %i"%i)

# use tight layout so texts don't overlap
plt.tight_layout()
```
